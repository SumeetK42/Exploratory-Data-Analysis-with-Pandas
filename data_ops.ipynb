# Data Profiling

```python
# Import libraries
# Load sample dataset (sample_data.csv)

df = pd.read_csv('sample.csv')

df
```

```python
## get the data type of columsn

data = df.info()

data
```

```python
# Total Row Count
# Calculate here

print ('Total Columns ',len(df.columns))
print ('Total Rows ',len(df.values))
```

```python
# Null Value Count
# Calculate for each column

for i in df.columns:
    print(i)
    print('Values count : ', df[i].count())
    print('Null values : ', df[i].isnull().sum())

```

```python
# Distinct Values
# Calculate for each column

for i in df.columns:
    print('Row Count : ',df[i].count())
    print('Unique values count : ',df[i].nunique())
    print('Unique Values :',df[i].unique())
```

```python
## to find the duplicate values in each columns

for i in df.columns:
    print('Columns' , i)
    print('Duplicate Values : ',df[i].duplicated().sum())
    print(df[i].value_counts()[df[i].value_counts() > 1])    
```

```python

# Min/Max/Avg/StdDev
# Calculate for numerical columns


stats_age =  {
   "avg age" :  df['age'].mean().__round__(),
   "median age" :  df['age'].median().__round__(),
   "Max age" :  df['age'].max().__round__(),
   "Min age" :  df['age'].min().__round__()
}

print(stats_age)

stats_purchase_amount =  {
   "avg purchase_amount " :  df['purchase_amount'].mean().__round__(),
   "median purchase_amount" :  df['purchase_amount'].median().__round__(),
   "Max purchase_amount" :  df['purchase_amount'].max().__round__(),
   "Min purchase_amount" :  df['purchase_amount'].min().__round__()
}

print(stats_purchase_amount)


##df['purchase_amount']

# Save Results
# Export to CSV or JSON
```
