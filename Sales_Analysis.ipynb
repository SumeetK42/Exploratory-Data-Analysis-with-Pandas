```python
import pandas as pd
import datetime as dt 
import time as t
import numpy as np
import openpyxl
from  matplotlib import pyplot as plt
```

```python
df5 = pd.read_pickle('df4.pkl')
```

```python
df5
```

```python
df5 = df5.rename(columns={"TotalPrice":"TotalSales"})
```

```python
df5.groupby('InvoiceYear')['TotalSales'].sum().plot(kind='bar')
plt.title('Year wise Sales')
plt.ylabel('Total Sale Amount')
plt.xlabel('Invoice Years')
plt.ylim(1000,20000000)
plt.yticks([100,1000,10000,100000,1000000,10000000,20000000])
plt.show()
```

```python
df5.sort_values('TotalSales').groupby('InvoiceMonth')['TotalSales'].sum().plot(kind='bar')
#plt.xticks(['Jan','Feb','Mar','Apr','May','June','Jul','Aug','Sept','Oct','Nov','Dec'])
```

```python
df5.groupby(['StockCode','Description'])['TotalSales'].sum().head(5).sort_values().plot(kind='barh')
plt.title('Product Wise Sales')
plt.xlabel('Stock description')
```

```python
df5.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
```
